```python
import pandas as pd
```

```python
#get country vaccine data
vaccine_df = pd.read_csv('country_vaccinations.csv')
vaccine_df.head()
```

```python
#drop columns that aren't needed
vaccine_df = vaccine_df.drop(['source_name','source_website','iso_code','daily_vaccinations_raw'], axis=1)
vaccine_df.head()
```

```python
#isolate the most recent data
max_date_df = pd.DataFrame(vaccine_df.groupby('country')['date'].max())
```

```python
cleaned_vaccine_df = pd.merge(max_date_df, vaccine_df, on = ['country', 'date'])
cleaned_vaccine_df
```
