## import dependecies

```python
import pandas as pd
```

## get country vaccine data

```python
vaccine_df = pd.read_csv('raw_csv/country_vaccinations.csv')
vaccine_df.head()
```

## drop columns that aren't needed from vaccine data

```python
vaccine_df = vaccine_df.drop(['source_name','source_website','iso_code','daily_vaccinations_raw'], axis=1)
vaccine_df.head()
```

## isolate the most recent date

```python
max_date_df = pd.DataFrame(vaccine_df.groupby('country')['date'].max())
```

```python
cleaned_vaccine_df = pd.merge(max_date_df, vaccine_df, on = ['country', 'date'])
cleaned_vaccine_df = cleaned_vaccine_df.set_index('country')
cleaned_vaccine_df.head()
```

## read in UN country data

```python
un_df = pd.read_csv("raw_csv/country_profile_variables.csv")
```

## drop columns that aren't needed from UN country data

```python
un_df = un_df.drop(['Surface area (km2)','GDP growth rate (annual %, const. 2005 prices)','Economy: Agriculture (% of GVA)','Economy: Industry (% of GVA)','Economy: Services and other activity (% of GVA)','Employment: Agriculture (% of employed)','Employment: Industry (% of employed)','Employment: Services (% of employed)','Labour force participation (female/male pop. %)','Agricultural production index (2004-2006=100)','Food production index (2004-2006=100)','International trade: Exports (million US$)','International trade: Imports (million US$)','International trade: Balance (million US$)','Balance of payments, current account (million US$)','Fertility rate, total (live births per woman)','International migrant stock (000/% of total pop.)','Refugees and others of concern to UNHCR (in thousands)','Infant mortality rate (per 1000 live births','Education: Government expenditure (% of GDP)','Education: Primary gross enrol. ratio (f/m per 100 pop.)','Education: Secondary gross enrol. ratio (f/m per 100 pop.)','Education: Tertiary gross enrol. ratio (f/m per 100 pop.)','Seats held by women in national parliaments %','Mobile-cellular subscriptions (per 100 inhabitants)','Mobile-cellular subscriptions (per 100 inhabitants).1','Individuals using the Internet (per 100 inhabitants)','Threatened species (number)','Forested area (% of land area)','CO2 emission estimates (million tons/tons per capita)','Energy production, primary (Petajoules)','Energy supply per capita (Gigajoules)','Pop. using improved drinking water (urban/rural, %)','Pop. using improved sanitation facilities (urban/rural, %)','Net Official Development Assist. received (% of GNI)'], axis=1)
un_df = un_df.set_index("country")
un_df.head()
```

## clean data where placeholder "-99" , ".../..." , "...", and "-~0.0"

```python
un_df = un_df.replace(-99,0)
un_df = un_df.replace(".../...",0)
un_df = un_df.replace("...",0)
un_df = un_df.replace("-~0.0",0)

un_df
```

## get world happiness data

```python
CSV_2019 = "raw_csv/2019.csv"

etl2019_df = pd.read_csv(CSV_2019)
etl2019_df
```

## drop columns that aren't needed from world happiness data

```python
etl2019_df = etl2019_df.drop(['Overall rank','GDP per capita'], axis = 1)
etl2019_df = etl2019_df.set_index("Country or region")

etl2019_df.head()
```

## write dataframes to sql

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:1316@localhost:5432/etl_world_data', paramstyle='format')
```

```python
cleaned_vaccine_df.to_sql('vaccine',engine, if_exists='replace')
```

```python
un_df.to_sql('UN',engine, if_exists='replace')
```

```python
etl2019_df.to_sql('happiness',engine, if_exists='replace')
```
